```python
class Atm:
  def __init__(self):
    self.pin=''
    self.balance=0
    print('Welcome to ATM')
    self.menu()
  def menu(self):
    user_input=input("""
    Hello,how would you like to proceed?
    1. Enter 1 to create pin
    2. Enter 2 to change pin
    3. Enter 3 to check balance
    4. Enter 4 to withdraw
    5. Enter 5 to exit
    """)
    if user_input=='1':
      #create pin
      self.create_pin()
    elif user_input=='2':
      #change pin
      self.change_pin()
    elif user_input=='3':
      #check balance
      self.check_balance()
    elif user_input=='4':
      #withdraw
      self.withdraw()
    else:
      print('Thank you so much for coming:)')
  def create_pin(self):
    user_pin=input("Enter your pin:")
    self.pin=user_pin
    user_balance=int(input('Enter your balance:₹'))
    self.balance=user_balance
    print('pin set successfully')
    self.menu()

  def change_pin(self):
    old_pin=input('Enter your old pin:')
    if old_pin==self.pin:
      #let him change the pin
      new_pin=input('Enter your new pin:')
      self.pin=new_pin
      print('pin changed successfully')
      self.menu()
    else:
      print('bhai teri pin galat hai')
      self.menu()

  def check_balance(self):
    user_pin=input('Enter your pin:')
    if user_pin==self.pin:
      user_balance=self.balance
      print('your balance is:₹',self.balance)
      self.menu()
    else:
      print('Bhai tu pin galat daal raha hai')
      self.menu()

  def withdraw(self):
    user_pin=input('Enter your pin:')
    if user_pin==self.pin:
      Amount=int(input("Enter the amount:₹"))
      if Amount<=self.balance:
        print('withdrawal successful')
        self.balance=self.balance-Amount
        self.menu()
      else:
        print('Bikhari Sale Pahale Paise to Jama kr Le')
        self.menu()
    else:
      print('Chore Sale Nikal Idhar Se')
      self.menu()

obj=Atm()
```
